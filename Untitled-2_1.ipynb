```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("prices.csv", sep = ";")
df
```

```python
df['product'].unique()
```

```python
a_df = df[df.get("product") == "AMETHYSTS"]
a_df
```

```python
s_df = df[df.get("product") == "STARFRUIT"]
s_df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(s_df['timestamp'], s_df['mid_price'], label='Mid Price', color='blue')
plt.title('Mid Price of STARFRUIT Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.show()
```
